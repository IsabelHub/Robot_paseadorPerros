```python
class Ruta:
    def __init__(self, nombre, calles, descripcion):
        self.nombre = nombre
        self.calles = calles
        self.descripcion = descripcion

    def __repr__(self):
        return f"{self.nombre}: {self.descripcion}"


def generar_rutas():
    ruta_A = Ruta(
        "Ruta A (verde)",
        ["Calle A", "Calle C", "Parque Central", "Calle E"],
        "Incluye parques, evita charcos, paseo relajado."
    )

    ruta_B = Ruta(
        "Ruta B (rápida)",
        ["Calle D", "Avenida Norte", "Calle F"],
        "Más directa y rápida, sin parques."
    )

    ruta_C = Ruta(
        "Ruta C (segura)",
        ["Calle A", "Calle B", "Calle Iluminada", "Calle E"],
        "Calles amplias y bien iluminadas."
    )

    return [ruta_A, ruta_B, ruta_C]


def elegir_ruta(rutas):
    print("Estas son las rutas disponibles:\n")
    for i, ruta in enumerate(rutas, start=1):
        print(f"{i}. {ruta.nombre} → {ruta.descripcion}")

    print("\n¿Cuál ruta quieres seguir? (1, 2 o 3)")

    while True:
        eleccion = input("Tu elección: ")

        if eleccion in ["1", "2", "3"]:
            return rutas[int(eleccion) - 1]
        else:
            print("No he entendido tu respuesta. Por favor elige 1, 2 o 3.")


rutas = generar_rutas()
ruta_elegida = elegir_ruta(rutas)

print("\nPerfecto, seguiremos esta ruta:")
print(f"- {ruta_elegida.nombre}")
print("Calles del recorrido:", ruta_elegida.calles)
```
